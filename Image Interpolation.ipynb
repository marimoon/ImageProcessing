```python
import numpy as np
from scipy import interpolate
import matplotlib
from PIL import Image
from matplotlib import pyplot as plt
%matplotlib inline
```

# Target Image

```python
num = 16  # the number of rings

x = np.arange( 0, 256 ) - 128
y = np.arange( 0, 256 ) - 128
xx, yy = np.meshgrid(x, y)
img = np.sin( np.pi*num*num*(xx**2+yy**2)/256/256)
plt.imshow(img)
```

# Interpolation function

You can select the interpolation method from `cubic`, `linear` and `quintic`.

```python
f_cubic = interpolate.interp2d( x, y, img, kind='cubic' )
f_linear = interpolate.interp2d( x, y, img, kind='linear' )
f_quintic = interpolate.interp2d( x, y, img, kind='quintic' )
```

Befor implementing the interpolation, the exact image is below.

```python
x2 = (np.arange( 0, 256 ) - 128) / 16
y2 = (np.arange( 0, 256 ) - 128) / 16
x2 += 96
y2 += 96
xx2, yy2 = np.meshgrid(x2, y2)
img_exact = np.sin( np.pi*num*num*(xx2**2+yy2**2)/256/256)
plt.imshow(img_exact)
```

Then, the interpolation functions are implemented.

```python
img_cubic = f_cubic(x2, y2)
img_linear = f_linear(x2, y2)
img_quintic = f_quintic(x2, y2)
```
